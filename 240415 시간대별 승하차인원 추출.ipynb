```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
df1 = pd.read_csv("./코로나 이전 승하차인원.csv")
df2 = pd.read_csv("./코로나 시기 승하차인원.csv")
df3 = pd.read_csv("./코로나 이후 승하차인원.csv")
```

##### [ 코로나 이전 ]

```python
df1.head()
```

```python
def categorize_weekday(요일):
    if 요일 in ['월', '화', '수', '목', '금']:
        return '평일'
    else:
        return '주말'
    
# '주간' 컬럼 생성
df1['주간'] = df1['요일'].apply(categorize_weekday)
```

```python
filtered_df1 = df1[df1['요일'].astype(str).str.contains('월|화|수|목|금')]
print(filtered_df1)
```

```python
# '07시-08시', '08시-09시', '09-10시 합쳐서 '출근시간대' 컬럼 생성
filtered_df1['출근시간대'] = filtered_df1['07시-08시'] + filtered_df1['08시-09시'] + filtered_df1['09시-10시']
filtered_df1.head()
```

```python
# '11시-12시', '12시-13시', '13시-14시', '14시-15시', '15시-16시' 합쳐서 '낮 시간대' 컬럼 생성
filtered_df1['낮 시간대'] = filtered_df1['11시-12시'] + filtered_df1['12시-13시'] + filtered_df1['13시-14시'] + filtered_df1['14시-15시'] + filtered_df1['15시-16시']
filtered_df1.head()
```

```python
# '17시-18시', '18시-19시', '19시-20시' 합쳐서 '퇴근시간대' 컬럼 생성
filtered_df1['퇴근시간대'] = filtered_df1['17시-18시'] + filtered_df1['18시-19시'] + filtered_df1['19시-20시']
filtered_df1.head()
```

```python
# '22시-23시', '23시-24시', '24시 이후' 합쳐서 '야간시간대' 컬럼 생성
filtered_df1['야간시간대'] = filtered_df1['22시-23시'] + filtered_df1['23시-24시'] + filtered_df1['24시 이후']
filtered_df1.head()
```
