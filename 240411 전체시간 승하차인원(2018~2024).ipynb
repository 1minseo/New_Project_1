```python
# '날짜' 컬럼을 datetime 형식으로 변환하여 다시 '날짜' 컬럼에 저장
df1['날짜'] = pd.to_datetime(df1['날짜'])
```

```python
# '날짜' 컬럼에서 year, momth, day 정보를 추출하여 '연도', '월', '일' 컬럼에 저장
df1['연도'] = df1['날짜'].dt.strftime("%Y")
df1['월'] = df1['날짜'].dt.strftime("%m")
df1['일'] = df1['날짜'].dt.strftime("%d")
```

```python
# 요일 정보를 숫자로 추출
df1['요일'] = df1['날짜'].dt.weekday
df1
```

```python
# 요일의 한글 이름을 추출
weekday_list = ['월', '화', '수', '목', '금', '토', '일']
df1['요일'] = df1.apply(lambda x: weekday_list[x['요일']], axis=1)
df1.head()
```

```python
# '분기' 컬럼 생성
def get_quarter(날짜):
    if 1 <= 날짜 <= 3:
        return 1
    elif 4 <= 날짜 <= 6:
        return 2
    elif 7 <= 날짜 <= 9:
        return 3
    else:
        return 4
    
df1['분기'] = df1['날짜'].dt.month.apply(get_quarter)
```

```python
df1.head()
```

```python
# '연번', '날짜', '역번호' 컬럼 제거
df1 = df1.drop(columns=['연번', '역번호'])
df1.head()
```

```python
# '뚝섬유원지'를 '자양'으로 변경
df1['역명'] = df1['역명'].replace('뚝섬유원지', '자양')
```

```python
# 괄호와 그 안의 내용을 삭제하고 수정된 값을 해당 컬럼에 할당

import re
df1['역명'] = df1['역명'].str.replace(r'\([^()]*\)', '', regex=True)
```

```python
# '역명' 컬럼에 있는 모든 단어 뒤에 '역'을 추가 ('서울역' 제외)
df1.loc[:, '역명'] = df1['역명'].apply(lambda x: x + '역' if x != '서울역' else x)
```

```python
# index를 포함하지 않고 데이터 저장
df1.to_csv("2024 승하차인원(1_8호선).csv", index=False)
```
