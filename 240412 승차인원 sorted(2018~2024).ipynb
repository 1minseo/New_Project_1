```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
df24 = pd.read_csv("./2024 승차인원(1_8호선).csv")
df23 = pd.read_csv("./2023 승차인원(1_8호선).csv")
df22 = pd.read_csv("./2022 승차인원(1_8호선).csv")
df21 = pd.read_csv("./2021 승차인원(1_8호선).csv")
df20 = pd.read_csv("./2020 승차인원(1_8호선).csv")
df19 = pd.read_csv("./2019 승차인원(1_8호선).csv")
df18 = pd.read_csv("./2018 승차인원(1_8호선).csv")
```

2024년

```python
df24
```

```python
# 각 분기별로 시간대별 승차 데이터를 큰 값부터 작은 값 순서로 sorted
df24_sorted =  df24.groupby('분기').apply(lambda x: x.sort_values(by=['06시 이전', '06시-07시', '07시-08시', '08시-09시', '09시-10시', '10시-11시',
                                                                    '11시-12시', '12시-13시', '13시-14시', '14시-15시', '15시-16시', '16시-17시',
                                                                    '17시-18시', '18시-19시', '19시-20시', '20시-21시', '21시-22시', '22시-23시',
                                                                    '23시-24시', '24시 이후'], ascending=False))
df24_sorted.head()
```

```python
# index를 포함하지 않고 데이터 저장
df24_sorted.to_csv("2024 분기별 승차인원 sorted.csv", index=False)
df24_sorted.head()
```
