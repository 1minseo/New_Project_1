2024 승차

```python
# 데이터 불러오기
df24 = pd.read_csv("./2024 승하차인원(1_8호선).csv")
df24.head()
```

```python
print(df24.dtypes)
```

```python
# '구분' 컬럼에서 '승차'가 포함된 값만 추출
df24_1 = df24[df24['구분'].str.contains('승차')]
```

```python
# index를 포함하지 않고 데이터 저장
df24_1.to_csv("2024 승차인원(1_8호선).csv", index=False)
df24_1.head()
```
