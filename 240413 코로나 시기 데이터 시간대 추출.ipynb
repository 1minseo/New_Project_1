```python
import pandas as pd
import numpy as np
```

```python
# 코로나 시기 데이터 불러오기
df = pd.read_csv("./코로나 시기 승하차인원.csv")
```

```python
df.head()
```

```python
def categorize_weekday(요일):
    if 요일 in ['월', '화', '수', '목', '금']:
        return '평일'
    else:
        return '주말'
    
# '주간' 컬럼 생성
df['주간'] = df['요일'].apply(categorize_weekday)    
```

```python
df.head()
```
