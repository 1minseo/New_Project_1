```python
df1
```

```python
print(df1.dtypes)
```

```python
# '날짜' 컬럼을 datetime 형식으로 변환하여 다시 '날짜' 컬럼에 저장
df1['날짜'] = pd.to_datetime(df1['날짜'])
```

```python
# '날짜' 컬럼에서 year 정보를 추출하여 '연도' 컬럼에 저장
df1['연도'] = df1['날짜'].dt.strftime("%y")
```

```python
# '분기' 컬럼 생성
def get_quarter(날짜):
    if 1 <= 날짜 <= 3:
        return 1
    elif 4 <= 날짜 <= 6:
        return 2
    elif 7 <= 날짜 <= 9:
        return 3
    else:
        return 4
    
df1['분기'] = df1['날짜'].dt.month.apply(get_quarter)
```

```python
df1.head()
```

```python
# '호선', '역명', '출근시간(07시-08시, 08시-09시)', '퇴근시간(17시-18시, 18시-19시)', '연도', '분기' 컬럼만 추출하기
df1_1 = df1[['호선', '역명', '07시-08시', '08시-09시', '17시-18시', '18시-19시', '연도', '분기']]
df1_1.head()
```

```python
# '뚝섬유원지'를 '자양'으로 변경
df1_1['역명'] = df1_1['역명'].replace('뚝섬유원지', '자양')
```

```python
# 괄호와 그 안의 내용을 삭제하고 수정된 값을 해당 컬럼에 할당

import regex
df1_1.loc[:, '역명'] = df1_1['역명'].str.replace(r'\(.*\)', '', regex=True)
```

```python
# '역명' 컬럼에 있는 모든 단어 뒤에 '역'을 추가 ('서울역' 제외)
df1_1.loc[:, '역명'] = df1_1['역명'].apply(lambda x: x + '역' if x != '서울역' else x)
```

```python
# index를 포함하지 않고 데이터 저장
df1_1.to_csv("2024 출퇴근시간 승하차인원(1_8호선).csv", index=False)
```
