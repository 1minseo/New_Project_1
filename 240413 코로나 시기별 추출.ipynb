```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
df24 = pd.read_csv("./2024 승하차인원_행정동 추가.csv")
df23 = pd.read_csv("./2023 승하차인원_행정동 추가.csv")
df22 = pd.read_csv("./2022 승하차인원_행정동 추가.csv")
df21 = pd.read_csv("./2021 승하차인원_행정동 추가.csv")
df20 = pd.read_csv("./2020 승하차인원_행정동 추가.csv")
df19 = pd.read_csv("./2019 승하차인원_행정동 추가.csv")
df18 = pd.read_csv("./2018 승하차인원_행정동 추가.csv")
```

코로나 시기 데이터 추출 (2020년 2분기 ~ 2022년 1분기)

```python
df20.head()
```

```python
print(df20['분기'].dtype)
```

```python
# 2020년 2분기~4분기 데이터 추출
filtered_df20 = df20[df20['분기'].astype(str).str.contains('2|3|4')]
print(filtered_df20)
```

```python
df22.head()
```

```python
# 2022년 1분기 데이터 추출
filtered_df22 = df22[df22['분기'].astype(str).str.contains('1')]
print(filtered_df22)
```

```python
dfs = [filtered_df20, df21, filtered_df22]
combined_df = pd.concat(dfs, ignore_index=True)
print(combined_df)
```

```python
# index를 포함하지 않고 데이터 저장
combined_df.to_csv("코로나 시기 승하차인원.csv", index=False)
```
