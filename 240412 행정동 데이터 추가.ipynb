```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
df24 = pd.read_csv("./2024 승하차인원(1_8호선).csv")
df23 = pd.read_csv("./2023 승하차인원(1_8호선).csv")
df22 = pd.read_csv("./2022 승하차인원(1_8호선).csv")
df21 = pd.read_csv("./2021 승하차인원(1_8호선).csv")
df20 = pd.read_csv("./2020 승하차인원(1_8호선).csv")
df19 = pd.read_csv("./2019 승하차인원(1_8호선).csv")
df18 = pd.read_csv("./2018 승하차인원(1_8호선).csv")
df = pd.read_excel("./지하철 주소 및 위경도(경기도 제외).xlsx")
```

```python
df.head()
```

```python
# '역번호', '호선', '역명' 컬럼 제거
df = df.drop(columns=['역번호', '호선', '역명'])
df.head()
```

```python
# '역명2' -> '역명'
df = df.rename(columns={'역명2':'역명'})
```

```python
# '뚝섬유원지'를 '자양'으로 변경
df['역명'] = df['역명'].replace('뚝섬유원지', '자양')
```

```python
# '역명' 컬럼에 있는 모든 단어 뒤에 '역'을 추가
df['역명'] = df['역명'].apply(lambda x: x + '역')
df.head()
```

2024년

```python
df24['역명'] = df24['호선'] + ' ' + df24['역명'].astype(str)
df24.head()
```

```python
# '호선' 컬럼 제거
df24 = df24.drop(columns=['호선'])
df24.head()
```

```python
df24_1 = pd.merge(df24, df, on = "역명")
df24_1.head()
```

```python
# index를 포함하지 않고 데이터 저장
df24_1.to_csv("2024 승하차인원_행정동 추가.csv", index=False)
```
